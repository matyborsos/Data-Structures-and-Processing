```python
from matplotlib import pyplot as plt
import numpy as np
fig,ax = plt.subplots(1,1)
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
ax.hist(a, bins = [0,25,50,75,100])
ax.set_title("Histogram")
ax.set_xticks([0,25,50,75,100])
ax.set_xlabel('grades')
ax.set_ylabel('number of students')
plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
ax.pie(students, labels = langs,autopct='%1.2f%%')
plt.show()
```

```python
import pandas as pd
import datetime
df=pd.DataFrame({'date_of_week': np.array([datetime.datetime(2023, 2, i+1)
                                                    for i in range(28)]),
                 'courses': np.random.randint(1, 15, size=28)})
plt.plot(df.date_of_week, df.courses)
plt.title('Classes by Date')
plt.xticks(rotation=30, ha='right')
plt.xlabel('Date')
plt.ylabel('Classes')
```

```python
def randrange(n, vmin, vmax):
    return (vmax - vmin) * np.random.rand(n) + vmin
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
for m, low, high in [('+', -50, -25), ('o', -30, -5)]:
    x=randrange(50, 23, 32)
    y=randrange(50, 0, 100)
    z=randrange(50, low, high)
    ax.scatter(x, y, z, marker=m)
plt.show()
```
